```python
import geopandas as gpd

m = folium.Map(location = [lati, long],tiles='openstreetmap',zoom_start = 12)
coords = []
coords2 = []
coords3 = []

for i in range(len(df1)-1):
    x = df1_x[i]
    y = df1_y[i]
    coords.append([x, y])
#시각화1
for i in range(len(coords)):
    folium.Circle(
        location = coords[i],
        radius = 50,
        color = '#FF69B4',
        fill = 'crimson',
    ).add_to(m)

for i in range(len(df2)-1):
    x_ = df2_x[i]
    y_ = df2_y[i]
    coords2.append([x_, y_])
#시각화2
for i in range(len(coords2)):
    folium.Circle(
        location = coords2[i],
        radius = 50,
        color = '#1E90FF',
        fill = 'crimson',
    ).add_to(m)

for i in range(len(df3)-1):
    x_ = df3_x[i]
    y_ = df3_y[i]
    coords3.append([x_, y_])
#시각화3
for i in range(len(coords3)):
    folium.Circle(
        location = coords3[i],
        radius = 50,
        color = '#8b00ff',
        fill = 'crimson',
    ).add_to(m)


#동별로 새로 만든 geojson 넣기
geojson_file = 'final.geojson'

style1 = {'fillColor': '#FFFFFF', 'color': '#e85255'}

test1=gpd.read_file(geojson_file)
folium.GeoJson(test1,name='test1',style_function=lambda x:style1).add_to(m)


#시각화
m
```
