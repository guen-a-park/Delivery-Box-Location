```python
import json
import sys
import pandas as pd
import requests
from datetime import *

# 전역 변수
accident_data = '/content/drive/MyDrive/공모전-Folium/gs+self-lat-lng.csv'
APP_KEY = ''    # 발급받은 키 입력
URL = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json'
```

```python
def json_request(url='', encoding='utf-8', success=None, error=lambda e: print('%s : %s' % (e, datetime.now()), file=sys.stderr)):
    headers = {'Authorization': 'KakaoAK {}'.format(APP_KEY)}
    resp = requests.get(url, headers=headers)
    print('%s : success for request [%s]' % (datetime.now(), url))

    return resp.text
```

```python
def reverse_geocode(longitude, latitude):
    # 파라미터 최적화하여 url 생성
    url = '%s?x=%s&y=%s' %(URL, longitude, latitude)

    # json request
    try:
        json_req = json_request(url=url)
        json_data = json.loads(json_req)
        json_doc = json_data.get('documents')[1]
        json_name = json_doc.get('region_1depth_name')
        json_name_2 =json_doc.get('region_2depth_name')
        json_name_3 =json_doc.get('region_3depth_name')
    except:
        json_name = 'NaN'
        json_name_2 = 'NaN'
        json_name_3 = 'NaN'

 

    return str(json_name)+' '+str(json_name_2)+' '+str(json_name_3)
```

```python
def get_address(data):
    address = []

    # 경도, 위도 추출해서 동 주소 반환
    for i in range(len(data)):
        longitude = data['lng'][i]
        latitude = data['lat'][i]
        address.append(reverse_geocode(longitude, latitude))

    return address  # 전처리 함수에서 주소 리스트 받아서 데이터프레임에 추가
```
