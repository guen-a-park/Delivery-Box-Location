```python
#csv 읽어오기
data = pd.read_csv('folium-example4.csv',engine='python',encoding='CP949')
data.head(5)
```

```python
#숫자 데이터 지도위에 시각화하기

# 서울시 중심부의 위도, 경도
center = [37.541, 126.986]

# zoom 레벨은 조정해 맵 만들기,tile 바꾸면 스타일 바뀜
#https://deparkes.co.uk/2016/06/10/folium-map-tiles/
population_map = folium.Map(location=center,tiles='Open street map', zoom_start=12)

# Choropleth 레이어를 만들기
population_map.choropleth(
    geo_data=geo_data,
    data=data,
    columns=['동', '1인가구수'],
    key_on='feature.properties.name',
    fill_color='BuPu',
    legend_name='number',
    smooth_factor=0)

# add tile layers to the map
tiles = ['stamenwatercolor', 'cartodbpositron', 'openstreetmap', 'stamenterrain']
for tile in tiles:
    folium.TileLayer(tile).add_to(population_map)

#tile 바꾸기 상자
folium.LayerControl().add_to(population_map)

# 맵을 출력
population_map
```

```python
import csv
import googlemaps

# 위치를 찾을 장소나 주소 지정 
# hwp 파일 참고해서 2021 기준 사용가능한곳 csv 만듦-> 데이터 정제과정 ppt 간략하게 설명하기

#csv 파일 list로 convert
f = open("location2.csv","rt")
csvfile = csv.reader(f)
places = []

for item in csvfile:
  places.append(item)

#2차원으로 저장돼서 1차원으로 차원축소
places = sum(places, [])
print(places)
```

```python
#발급받은 api key 넣기
my_key = ""
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도

i=0
for place in places:   
    i = i + 1
    try:
        print("%d번 인덱스에서 %s의 위치를 찾고있습니다"%(i, place))
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))


# 데이터프레임만들어 출력하기
df = pd.DataFrame({'lat':lat, 'lng':lng}, index=places)
#df.head(5)
```

```python
#결측치 google map 검색 후 직접수정. 대부분 지명이 조금 변경거나 다른 동일지명 때문에 오류

#33,35,46,48,74,80,108,111,137,192,193,204 인덱스 잘못됨

df.iloc[[33,35]] = [(37.53830896854011, 127.0878270398395),(37.547209307462786, 127.09005508216892)]
df.iloc[[33,35,46,48,74,80,108,111,137,192,193,204]] = [(37.53830896854011, 127.0878270398395),(37.547209307462786, 127.09005508216892),(37.59438189471257, 127.05968719751229),
(37.580618074547445, 127.05303742883679),(37.636195779723074, 127.02993871665113),(37.651912363204005, 127.03729445522251),(37.61832452069178, 126.93300424567866),
(37.57623598949008, 126.92638632143401),(37.51900504299165, 126.85378777052264),(37.525962769539674, 126.89396033828044),(37.478865490490385, 126.97323508213337),(37.4864472440814, 126.90767106411226)]

#시각화시 결측치 더 찾음
df.iloc[[198,207,215]] = [(37.49484422691444, 126.93081725561063),(37.4758099993684, 126.93787855375372),(37.479153590986954, 126.94154816724404)]

print(df.iloc[215])

#index 162없애기
df = df.drop(df.index[162])

df.shape
```

```python
from pyproj import Transformer
import pyproj
from shapely.geometry import Point as point
#import geopandas as gpd
```

```python
# csv 불러오기
parcel = pd.read_csv('lat-lng.csv',engine='python') 
parcel_x = parcel['lat']
parcel_y = parcel['lng']
```

```python
#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
lati = parcel['lat'].mean()
long = parcel['lng'].mean()

print(lati)
print(long)
# #지도 띄우기
# m = folium.Map(location = [lati, long],tiles="cartodbpositron",zoom_start = 12)
# m
```

```python
m = folium.Map(location = [lati, long],tiles="cartodbpositron",zoom_start = 12)
coords = []
for i in range(len(parcel)-1):
    x = parcel_x[i]
    y = parcel_y[i]
    coords.append([x, y])

#시각화 설정바꾸기, 마크 바꾸기?
for i in range(len(coords)):
    folium.Circle(
        location = coords[i],
        radius = 50,
        color = '#8b00ff',
        fill = 'crimson',
    ).add_to(m)

#시각화
m
```
